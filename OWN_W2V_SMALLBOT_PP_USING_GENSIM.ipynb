```python
import os
import json
import nltk
import gensim
import numpy as np
from gensim import corpora, models, similarities
import pickle
```

```python
#loading Json conversation file
file=open('conversation.json',);
data = json.load(file)
```

```python
#printing the data
#data
```

```python
#savind the only conversation data in the cor object
cor=data["conversations"];
```

```python
#checking the length of cor
len(cor)
```

```python
#spliting the Que and Ans into two variables x,y
x=[]
y=[]

for i in range(len(cor)):
    for j in range(len(cor[i])):
        if j<len(cor[i])-1:
            x.append(cor[i][j]);
            y.append(cor[i][j+1]);
```

```python
#text cleaning
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r"i'm", "i am", text)
    text = re.sub(r"he's", "he is", text)
    text = re.sub(r"it's", "it is", text)
    text = re.sub(r"It's", "it is", text)
    text = re.sub(r"she's", "she is", text)
    text = re.sub(r"that's", "that is", text)
    text = re.sub(r"what'd", "what did", text)
    text = re.sub(r"that's", "that is", text)
    text = re.sub(r"didn't", "did not", text)
    text = re.sub(r"this'll", "this will", text)
    text = re.sub(r"you're", "you are", text)
    text = re.sub(r"It's", "it is", text)
    text = re.sub(r"don't", "do not", text)
    text = re.sub(r"he'll", "he will", text)
    text = re.sub(r"can't", "can not", text)
    text = re.sub(r"who's", "who is", text)
    text = re.sub(r"there's", "there is", text)
    text = re.sub(r"why'd", "why did", text)
    text = re.sub(r"They'll", "They will", text)
    text = re.sub(r"couldn't", "could not", text)    
    text = re.sub(r"what's", "what is", text)
    text = re.sub(r"how's", "how is", text)
    text = re.sub(r"where's", "where is", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"n't", "not", text)
    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"'bout", "about", text)
    text = re.sub(r"'till", "untill", text)
    text = re.sub(r"n'", "ng", text)
    text = re.sub(r"[-()\"#/@;:<>{}+=~|.?,]", "", text)
    return text
```

```python
#cleaning the question
clean_x=[]
for i in x:
    clean_x.append(clean_text(i))
```

```python
#cleaning the answers
clean_y=[]
for i in y:
    clean_y.append(clean_text(i))
```

```python
# tokenizing the ans and que
tok_x=[]
tok_y=[]
for i in range(len(clean_x)):
    tok_x.append(nltk.word_tokenize(clean_x[i].lower()))
    tok_y.append(nltk.word_tokenize(clean_y[i].lower()))
```

```python
# checking for tokenization
tok_x[1:5]
```

```python
## removing stopwords
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
stop_words = set(stopwords.words('english'))  
```

```python
#removing stopwords from que
for i in tok_x: 
    for w in i:
        if w in stop_words: 
            i.remove(w) 
```

```python
#removing stopwords fron answers
for i in tok_y: 
    for w in i:
        if w in stop_words: 
            i.remove(w) 
```
